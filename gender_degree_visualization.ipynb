```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
```

```python
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
cb_light_gray = (171/255, 171/255, 171/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
major_cats = [stem_cats, lib_arts_cats, other_cats]
cat_names = ['STEM', 'Humanities', 'Other']

fig = plt.figure(figsize=(18, 18))
fig.suptitle('Gender Percentage by Degree', fontsize=20)

for col_i, cat in enumerate(major_cats):
    for row_i, major in enumerate(cat):
        major = cat[row_i]
        ax = fig.add_subplot(6,3,col_i+3*row_i+1)
        ax.plot(women_degrees['Year'], women_degrees[major], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[major], c=cb_orange, label='Men', linewidth=3)
        for key, spine in ax.spines.items():
            spine.set_visible(False)
        ax.set_xlim(1968, 2011)
        ax.set_ylim(0,100)
        ax.set_title(major)
        ax.set_yticks([0, 100])
        ax.axhline(50, c=cb_light_gray, alpha=.3)
        ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)

        if row_i == 0:
            ax.set_title('{}\n\n\n{}'.format(cat_names[col_i], major))
            ax.text(2005, 87, 'Men')
            ax.text(2002, 8, 'Women')
        elif row_i + 1 == len(cat):
            ax.text(2005, 62, 'Men')
            ax.text(2001, 35, 'Women')
            ax.tick_params(labelbottom=True)
plt.savefig('gender_degrees.png')
plt.show()
```
